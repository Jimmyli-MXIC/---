```python
import csv
from bs4 import BeautifulSoup
#打开文件，用with打开可以不用去特意关闭file了，python3不支持file()打开文件，只能用open()
with open("/home/jimmyli/bulletinu.csv","r",encoding="utf-8") as csvfile:
     #读取csv文件，返回的是迭代类型

    
    dictread=csv.DictReader(csvfile)
    
    for i ,row in enumerate(dictread):
        if(i<833):
            print("\n\n===============================\n[ "+ str(i) +" ]\n",row)
            html = row['公告内容']
            bsObj= BeautifulSoup(html,"lxml")
            #print(html)
            print("\n== 公告内容 gettext\n",bsObj.get_text())
            imgs = bsObj.findAll('img')
            for img in imgs:
                print("\n==== img.src:\n",img.get('src'))
```

```python
import multiprocessing
import jieba

tmp_catalog = '/home/jimmyli/announcement/'
file_list = [tmp_catalog+'train.txt', tmp_catalog+'test.txt']
write_list = [tmp_catalog+'train_token.txt', tmp_catalog+'test_token.txt']

def tokenFile(file_path, write_path):
    token_sen = []
    with open(write_path, 'w') as w:
        with open(file_path, 'r') as f:
            for line in f:
                token_sen = (" ".join(jieba.cut(line.split('\t')[1])))
                w.write(line.split('\t')[0].encode('utf-8') + '\t' + token_sen.encode('utf-8') + '\n')
    print(file_path + ' has been token and token_file_name is ' + write_path)

tokenFile(tmp_catalog+'train.txt', tmp_catalog+'train_token.txt')
```
